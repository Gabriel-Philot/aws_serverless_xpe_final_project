```python
import requests
import pandas as pd
import boto3
import json
import datetime
import pyarrow as pa
import pyarrow.parquet as pq
import io
```

```python
url = "https://api.infomoney.com.br/fii/0/1000/36/desc/imovel/null?type=json"

def web_screp_info_money(url):
  """
  Requests of the web screp Infomoney.

 Return: json_file
  """

  try:
    response = requests.get(url)
    
    if response.status_code != 200:
      return False
    
    data = response.json()
    
    return data

  except:
    return False
  

json_data = web_screp_info_money(url)
```

```python
df = pd.json_normalize(json_data)


bucket = "philot-bucket-lambda0"
prefix = "landing/"+"test_info_money"

prof_name ="bilbo-lambda-test0-user"


print("Aqui")

table = pa.Table.from_pandas(df)
parquet_buffer = io.BytesIO()
pq.write_table(table, parquet_buffer, compression='snappy')
#session = boto3.Session(profile_name=prof_name)
#s3_resource = boto3.resource('s3')
#s3_resource.Object(bucket, prefix).put(Body=parquet_buffer.getvalue())

print("foi")
```
